```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the GeoJSON data
geojson_path = "h3_0_shifted.geojson"
gdf = gpd.read_file(geojson_path)

# Plot using matplotlib
fig, ax = plt.subplots(figsize=(10, 5))
crs = 'proj=moll'
crs = 'EPSG:3857'
gdf.to_crs(crs).plot(ax=ax, color="lightblue", edgecolor="black")
ax.set_title("Antimeridian GeoJSON")
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.set_aspect('equal', adjustable='box')
plt.show()
```

```python
import leafmap

# Choose projection for Leafmap rendering (Mollweide by default)
leafmap_projection = "EPSG:4326"  # Mollweide projection
gdf_leafmap = gdf.to_crs(leafmap_projection) if leafmap_projection else gdf

leafmap_center = [gdf.geometry.centroid.y.mean(), gdf.geometry.centroid.x.mean()]
m = leafmap.Map(center=leafmap_center, zoom=3)
m.add_gdf(gdf_leafmap, layer_name="Antimeridian", style=dict(color="blue", fillColor="cyan", fillOpacity=0.3))
m
```
